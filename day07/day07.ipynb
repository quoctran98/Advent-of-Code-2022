### Part 1

Find all of the directories with a total size of at most 100000. What is the sum of the total sizes of those directories?

```python
files = {}
current_dir = []

for line in input:
    line = line.split(" ")
    if line[0] == "$":
        if line[1] == "cd":
            if line[2] == "/":
                current_dir = []
            elif line[2] == "..":
                current_dir.pop()
            else:   
                current_dir.append(line[2])
    else: # populate files
        if line[0] != "dir":
            filepath = current_dir.copy()
            filepath.append(line[1])
            filepath = "/".join(filepath)
            files[filepath] = line[0]

```

```python
dirs = [f.split("/") for f in files.keys()]
dirs = set([tuple(d[:len(d)-1]) for d in dirs])

answer = 0

for this_dir in dirs:
    dir_size = 0
    for filepath, filesize in files.items():
        filepath = tuple(filepath.split("/")[:len(this_dir)])
        # if leading elements match up to this_dir then within dir
        if filepath == this_dir:
            dir_size += int(filesize)
    if dir_size <= 100000:
        answer += dir_size
    if dir_size == 0:
        print("ugh")
```
